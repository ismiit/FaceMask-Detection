```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
import tensorflow as tf
import keras
import sys
import os
import random
import datetime
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
from sklearn.metrics import accuracy_score, classification_report
from keras.models import Sequential
from keras.layers import Conv2D, MaxPool2D, Dense, Dropout, BatchNormalization, Flatten, AveragePooling2D
from keras import Input
from keras.datasets import fashion_mnist
from tensorflow.keras.utils import to_categorical
from keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.applications.vgg19 import VGG19
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import LabelBinarizer
```

```python
main_dir = os.path.dirname(sys.path[0])
main_dir
```

```python
family_link = main_dir+'/recognize_family/data/family/'
family_link
```

```python
labels = set(['with_mask','without_mask'])
```

```python
image_paths = []
image_labels = []

for label in labels:
    for image in list(os.listdir(family_link+label)):
        image_paths = image_paths + [family_link + label + '/' +image]
        image_labels = image_labels + [label]
```

```python
np.shape(image_labels)
```

```python
image_paths
```

```python
image_labels
```

```python
data = []
for img in image_paths:
    image = cv2.imread(img)
    image = cv2.resize(image,(224,224))
    data.append(image)
```

```python
X = np.array(data,dtype='float')
```

```python
X
```

```python
y = np.array(image_labels)
y
```

```python
print('shape of labels :',y.shape)
print('shape of dataset :',X.shape)
```

```python
lb = LabelBinarizer()
y = lb.fit_transform(y.reshape(-1,1))
y = to_categorical(y)
```

```python
np.unique(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
def display_image(X,y):
    fig, ax = plt.subplots(5,5,sharey=False,tight_layout=True, figsize=(12,12), facecolor='white')
    for i in range(0,5):
        for j in range(0,5):
            n = np.random.randint(0,len(y))
            ax[i,j].matshow(X[n])
            ax[i,j].set(title = y[n])
            ax[i,j].axis('off')
    plt.show()
```

```python
INIT_LR = 1e-4
EPOCHS = 20
BS = 32
```

```python
aug = ImageDataGenerator(rotation_range=20,zoom_range=0.15,width_shift_range=0.2,height_shift_range=0.2,shear_range=0.15,horizontal_flip=True,fill_mode='nearest')
```

```python
base_model = MobileNetV2(weights='imagenet',include_top=False,input_tensor=Input(shape=(224,224,3)))
```

```python
head_model = base_model.output
head_model = AveragePooling2D(pool_size=(2,2))(head_model)
head_model = Flatten(name='Flatten')(head_model)
head_model = Dense(128,activation='relu')(head_model)
head_model = Dropout(0.5)(head_model)
head_model = Dense(2,activation='softmax')(head_model)
```

```python
model = Model(inputs=base_model.input,outputs=head_model)
```

```python
for layer in base_model.layers:
    layer.trainable = False
```

```python
opt = Adam(learning_rate=INIT_LR, decay=INIT_LR / EPOCHS)
```

```python
model.compile(optimizer=opt,loss="binary_crossentropy",metrics=['accuracy'])
```

```python
model.summary()
```

```python
model_run = model.fit(aug.flow(X_train, y_train, batch_size=BS),
                      steps_per_epoch=len(X_train) // BS,validation_data=(X_test, y_test),validation_steps=len(X_test) // BS,epochs=EPOCHS)
```

```python
loss, acc = model.evaluate(X_test,y_test)
```

```python
print('loss : ',loss)
print('accuracy : ',acc*100)
```

```python
preds = model.predict(X_test)
```

```python
y_preds = np.argmax(preds,axis=1)
```
